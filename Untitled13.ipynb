Escape characters are special characters in a string that are used to represent characters that cannot be typed or printed normally. Escape characters are represented with a backslash (\) followed by another character.

Here are some commonly used escape characters in Python:

\n: represents a new line
\t: represents a tab
\': represents a single quote
\": represents a double quote
\\: represents a backslash

#answer2
The escape character \n stands for a new line, and the escape character \t stands for a horizontal tab.
The \n escape character is used to insert a line break, causing the text after the escape character to start on a new line. This is useful when you want to split text into multiple lines or create paragraphs.
The \t escape character is used to insert a horizontal tab, which creates a space of a certain width. This is useful when you want to align text or create columns.

#answer3
To include a backslash character in a string, you need to use an escape character (\) before the backslash. This is because the backslash is normally used as an escape character itself, so if you want to include a literal backslash in a string, you need to escape it with another backslash.

In Python, you can enclose a string in either single quotes (') or double quotes ("), as long as the starting and ending quotes match. This means that you can include a single quote character within a string that is enclosed in double quotes, or vice versa, without needing to escape it.

So in the string "Howl's Moving Castle", the single quote character within the word "Howl's" does not need to be escaped because the entire string is enclosed in double quotes. Python understands that the single quote character within the string is a part of the string itself, and not the end of the string.

In Python, there are three methods that can be used to adjust the alignment of a string within a specified width:

ljust(width[, fillchar]): This method left-justifies the string within the specified width. It pads the end of the string with the specified fillchar (if provided) or a space character (by default) to reach the given width.

Example: 'Hello'.ljust(10) returns the string 'Hello '

rjust(width[, fillchar]): This method right-justifies the string within the specified width. It pads the beginning of the string with the specified fillchar (if provided) or a space character (by default) to reach the given width.

Example: 'Hello'.rjust(10) returns the string ' Hello'

center(width[, fillchar]): This method centers the string within the specified width. It pads both sides of the string with the specified fillchar (if provided) or a space character (by default) to reach the given width, with the original string centered.

Example: 'Hello'.center(10) returns the string ' Hello '

In Python, there are two methods that can be used to remove whitespace characters from the start or end of a string:

lstrip([chars]): This method removes whitespace characters (or the specified chars characters) from the left side (start) of the string.

Example: ' Hello '.lstrip() returns the string 'Hello '

rstrip([chars]): This method removes whitespace characters (or the specified chars characters) from the right side (end) of the string.

Example: ' Hello '.rstrip() returns the string ' Hello'
