#Q1 In what modes should the PdfFileReader() and PdfFileWriter() File objects will be opened?
When working with the PyPDF2 library in Python, you should open the PdfFileReader() and PdfFileWriter() file objects in binary mode ('rb' for reading and 'wb' for writing).

Here's an example:
import PyPDF2

# Open the PDF file in read-binary mode
with open('example.pdf', 'rb') as pdf_file:
    # Create a PdfFileReader object
    pdf_reader = PyPDF2.PdfFileReader(pdf_file)
    # Do something with the PdfFileReader object

# Open a new PDF file in write-binary mode
with open('output.pdf', 'wb') as output_file:
    # Create a PdfFileWriter object
    pdf_writer = PyPDF2.PdfFileWriter()
    # Do something with the PdfFileWriter object
    # (e.g. add pages to the PDF, write the PDF to disk, etc.)
    pdf_writer.write(output_file)
Note that when working with PyPDF2, it's important to open the PDF file in read-binary mode, even if you're only reading text from the file. This is because PDF files are binary files and contain non-textual data, such as images and fonts, that can be corrupted if the file is read as a text file.

#Q4 If a PdfFileReader object’s PDF is encrypted with the password swordfish, what must you do
before you can obtain Page objects from it?
   If a PdfFileReader object's PDF is encrypted with a password, you must decrypt the PDF before you can obtain Page objects from it. In this case, you can use the decrypt() method of the PdfFileReader object to decrypt the PDF by passing the password as an argument.

Here's an example:
import PyPDF2

# Open the encrypted PDF file in read-binary mode
with open('example_encrypted.pdf', 'rb') as pdf_file:
    # Create a PdfFileReader object
    pdf_reader = PyPDF2.PdfFileReader(pdf_file)
    
    # Decrypt the PDF using the password 'swordfish'
    pdf_reader.decrypt('swordfish')
    
    # Get a Page object for page 5
    page_5 = pdf_reader.getPage(4)
    
    # Do something with the Page object (e.g. extract text from the page)
    page_text = page_5.extractText()
    
    # Print the text from page 5
    print(page_text)
In this example, we open the encrypted PDF file 'example_encrypted.pdf' in read-binary mode and create a PdfFileReader object called pdf_reader. We then use the decrypt() method of the pdf_reader object to decrypt the PDF using the password 'swordfish'. After the PDF is decrypted, we can retrieve the Page object for page 5 using pdf_reader.getPage(4) and perform operations on it as needed.

#Q6 What is the difference between a Run object and a Paragraph object?
   In Python's python-docx library, a Run object and a Paragraph object are two different types of objects that are used to represent different elements in a Word document.
A Paragraph object represents a single paragraph in a Word document. It contains a list of one or more Run objects that represent the formatted text within the paragraph. A Paragraph object may contain multiple Run objects if the text within the paragraph has different formatting applied to different sections.
A Run object, on the other hand, represents a contiguous run of text within a paragraph that has the same formatting applied to it. For example, if a paragraph contains the text "This is some bold text.", the paragraph would contain a single Run object for the entire text string. However, if the same text string was formatted to have the first three words in bold and the last two words in italics, then the paragraph would contain two Run objects - one for the bold text, and one for the italicized text.
In summary, a Paragraph object represents a single paragraph, and contains one or more Run objects that represent the formatted text within the paragraph. A Run object represents a contiguous run of text within a paragraph that has the same formatting applied to it.

#Q7 How do you obtain a list of Paragraph objects for a Document object that’s stored in a variable
named doc?
To obtain a list of Paragraph objects for a Document object stored in a variable named doc using the python-docx library, you can use the paragraphs attribute of the Document object. The paragraphs attribute returns a list of all the Paragraph objects in the document.

Here's an example:

import docx

# Open the Word document in read mode
doc = docx.Document('example.docx')

# Get a list of all the Paragraph objects in the document
paragraphs = doc.paragraphs

# Print the text of each paragraph in the document
for paragraph in paragraphs:
    print(paragraph.text)
