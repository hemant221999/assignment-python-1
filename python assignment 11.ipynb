#Q1 Create an assert statement that throws an AssertionError if the variable spam is a negative
integer.
answer-Here's an assert statement that throws an AssertionError if the variable spam is a negative integer:
assert spam >= 0, "spam cannot be a negative integer"
When this statement is executed, Python checks whether the condition spam >= 0 is true. If spam is a non-negative integer, the condition is true and the program continues executing. But if spam is a negative integer, the condition is false, and Python raises an AssertionError with the message "spam cannot be a negative integer".

#Q4 What are the two lines that must be present in your software in order to call logging.debug()?
answer-In order to call logging.debug() in your software, you must have the following two lines present:

import logging
logging.basicConfig(level=logging.DEBUG, format='%(asctime)s - %(levelname)s - %(message)s')
The first line imports the logging module, which provides the debug() method for logging debug messages.

The second line sets up the logging configuration, including the minimum logging level (logging.DEBUG) and the format of the log messages ('%(asctime)s - %(levelname)s - %(message)s'). You can customize the format string to include additional information in the log messages, such as the name of the module or function that generated the message.

With these two lines in place, you can call logging.debug() in your code to log debug messages.

#Q5 What are the two lines that your program must have in order to have logging.debug() send a
logging message to a file named programLog.txt?
answer-To configure logging.debug() to send log messages to a file named programLog.txt, you can use the following two lines of code:

import logging

logging.basicConfig(filename='programLog.txt', level=logging.DEBUG, format='%(asctime)s - %(levelname)s - %(message)s')
The first line imports the logging module.

The second line configures the logging module to write log messages to a file named programLog.txt. The filename parameter specifies the name of the file to write to. The level parameter sets the minimum logging level to logging.DEBUG, so that all messages with a severity of DEBUG and higher will be logged. The format parameter specifies the format of the log messages.

With these two lines in place, you can call logging.debug() in your code to write log messages to the file programLog.txt.

#Q6 What are the five levels of logging?
answer- DEBUG: Detailed information, typically of interest only when diagnosing problems.
INFO: Confirmation that things are working as expected.
WARNING: An indication that something unexpected happened or indicative of some problem in the near future (e.g., 'disk space low'). The software is still working as expected.
ERROR: Due to a more serious problem, the software has not been able to perform some function.
CRITICAL: A very serious error, indicating that the program itself may be unable to continue running.

#Q7 What line of code would you add to your software to disable all logging messages?
answer- To disable all logging messages in your software, you can add the following line of code:
logging.disable(logging.CRITICAL)
This line sets the logging level to CRITICAL, which effectively disables all logging messages with a severity level lower than or equal to CRITICAL. This means that DEBUG, INFO, WARNING, and ERROR messages will be suppressed, and only CRITICAL messages (such as those generated by a call to logging.critical()) will be allowed to propagate.


#Q8 Why is using logging messages better than using print() to display the same message?
answer- Using logging messages is generally better than using print() to display the same message for several reasons:

Configurability: Logging messages can be easily configured to be turned on or off, or to be sent to different destinations, such as a file, a database, or a remote server. This provides more flexibility and control over the output of your program, which can be especially useful in production environments.

Granularity: Logging messages can be categorized by severity level, allowing you to filter and analyze messages based on their importance. This makes it easier to prioritize and troubleshoot issues, and to identify patterns or trends in the behavior of your program.

Consistency: Logging messages provide a standardized way of reporting errors, warnings, and other messages, making it easier for developers and users to understand and interpret the output of your program. In contrast, print() statements can be scattered throughout your code and may not follow a consistent format or style.

Performance: Logging messages can be more efficient than print() statements, especially if you need to write large amounts of output to a file or database. The logging module is optimized for performance and provides features such as buffering and asynchronous logging to help reduce overhead.

Overall, using logging messages provides a more robust and flexible way of handling program output than using print(), especially in larger and more complex programs.

#Q9 What are the differences between the Step Over, Step In, and Step Out buttons in the debugger?
answer- Step Over: When you click the Step Over button, the debugger executes the current line of code and moves to the next line. If the current line of code contains a function call, the debugger will execute the function and return to the next line of code in the current function. This is useful for quickly stepping through a program without getting bogged down in details.

Step In: When you click the Step In button, the debugger enters the function call on the current line of code, and stops at the first line of code inside the function. This is useful for debugging functions and for tracing the flow of execution inside a program.

Step Out: When you click the Step Out button, the debugger executes the remaining lines of code inside the current function and returns to the line of code that called the function. This is useful for quickly exiting a function and returning to the main flow of execution.

#Q10 After you click Continue, when will the debugger stop ?
answer- When you click the Continue button in the debugger, the debugger will continue executing the program without interruption until one of the following occurs:
A breakpoint is encountered: If you have set a breakpoint in your code, the debugger will stop at the line of code that corresponds to the breakpoint.
An exception is raised: If your program raises an exception, the debugger will stop at the line of code where the exception occurred, allowing you to examine the traceback and diagnose the error.
The program finishes executing: If your program completes its execution without encountering any errors or breakpoints, the debugger will stop when the program finishes.
In general, the Continue button is used to allow your program to run uninterrupted until a specific condition is met, such as reaching a breakpoint or encountering an error. By allowing your program to run without interruption, you can get a better understanding of how it behaves in real-world scenarios, and can more easily identify and diagnose errors that might be difficult to reproduce under controlled conditions.

#Q11 What is the concept of a breakpoint?
answer-In software development, a breakpoint is a point in your code where the debugger is instructed to pause program execution, allowing you to examine the current state of the program, including the values of variables, the call stack, and other relevant information.
By setting a breakpoint at a specific line of code, you can pause program execution at that point and examine the current state of the program to better understand how it is behaving. This can be useful for diagnosing errors, testing hypotheses about program behavior, and understanding the flow of execution through your code.
breakpoints can be set in a variety of ways, including through the use of an integrated development environment (IDE), command-line tools, or specialized debugging libraries. Once a breakpoint is set, the debugger will pause program execution at that point, allowing you to interact with the program and examine its state. You can then step through the code one line at a time, evaluate expressions, and modify variables as needed to better understand the program's behavior.
Overall, breakpoints are a powerful tool for debugging software, allowing you to gain insights into how your code is executing and helping you to diagnose and fix errors more quickly and efficiently.
