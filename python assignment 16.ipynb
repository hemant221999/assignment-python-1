#Q10 Make a French-to-English dictionary called f2e from e2f. Use the items method.
answer-  Here's how you can create the French-to-English dictionary "f2e" from the existing English-to-French dictionary "e2f" using the items() method:
e2f = {
    'dog': 'chien',
    'cat': 'chat',
    'walrus': 'morse'
}

f2e = {value: key for key, value in e2f.items()}

print(f2e)
