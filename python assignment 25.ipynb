#Q1 What is the difference between enclosing a list comprehension in square brackets andparentheses?
Answer- The difference between enclosing a list comprehension in square brackets ([]) and parentheses (()) lies in the type of object that is created.

Square Brackets ([]):
When a list comprehension is enclosed in square brackets, it creates a new list object. The result of the list comprehension is a list that contains the elements generated by the comprehension.
Parentheses (()):
When a list comprehension is enclosed in parentheses, it creates a generator object or a generator expression. The result is a generator, which is an iterator that generates the values on-the-fly as they are needed.
#Q2 What is the relationship between generators and iterators?
Answer- Generators and iterators are closely related concepts in Python. In fact, generators are a type of iterator.

An iterator is an object that implements the iterator protocol, which consists of the __iter__() and __next__() methods. The __iter__() method returns the iterator object itself, and the __next__() method returns the next item from the iterator. If there are no more items, it raises the StopIteration exception.

Generators, on the other hand, are a convenient way to create iterators. They are functions that use the yield keyword to produce a sequence of values. When a generator function is called, it returns a generator object, which is an iterator. The generator object can be used to iterate over the values generated by the generator function.

#Q3 What are the signs that a function is a generator function?
Answer- Presence of the yield keyword: The primary indicator of a generator function is the presence of the yield keyword in the function body. The yield statement is used to produce a value and temporarily suspend the function's execution, allowing it to be resumed later to generate the next value.

Use of the return statement: Generator functions may use the return statement, but it behaves differently compared to regular functions. In a generator function, return is used to indicate the end of the generator and will raise a StopIteration exception when called again. However, the primary mechanism for generating values in a generator function is the yield statement.

Function type: Generator functions return a generator object when called, rather than directly returning a value or a different object. The generator object is an iterator that can be used to iterate over the values generated by the generator function.

#Q4 What is the purpose of a yield statement?
Answer- The yield statement is a powerful feature in Python that is used in the context of generator functions. It serves two main purposes:

Value Production: The primary purpose of the yield statement is to produce a value and temporarily suspend the execution of the generator function. When a yield statement is encountered, it behaves as a generator expression, returning the yielded value to the caller.
Function State Preservation: The yield statement also preserves the state of the generator function, allowing it to resume execution from where it left off when called again. This behavior distinguishes generator functions from regular functions, which start execution from the beginning every time they are called.
The yield statement allows generator functions to generate a sequence of values lazily, producing values only when needed. This enables efficient memory usage, as values are generated on-the-fly rather than generating and storing all values upfront. The ability to suspend and resume execution also provides a powerful mechanism for writing iterative code that operates on large or infinite sequences.

#Q5 What is the relationship between map calls and list comprehensions? Make a comparison and
contrast between the two.
Answer- The relationship between map() calls and list comprehensions in Python is that they both provide ways to transform and process iterable objects. However, they differ in their syntax and usage. Here's a comparison and contrast between map() calls and list comprehensions:

Purpose:

map(): The map() function applies a given function to each item of an iterable and returns an iterator that yields the results. It is primarily used to transform elements of one or more iterables using a specified function.
List comprehensions: List comprehensions provide a concise way to create new lists by applying an expression to each element of an iterable, optionally with filtering based on conditions. They are used to generate new lists based on existing lists or other iterable objects.
Syntax:

map(): The syntax of a map() call is map(function, iterable), where function is the function to be applied to each element of iterable. The result is an iterator that yields the transformed values.
List comprehensions: The syntax of a list comprehension is [expression for item in iterable if condition], where expression is the transformation or evaluation to be applied to each item from the iterable. The optional condition allows filtering based on specified criteria.
Return Type:

map(): The map() function returns an iterator, which can be converted to other types such as a list using the list() function if desired.
List comprehensions: List comprehensions directly create a new list containing the transformed elements, eliminating the need for an explicit conversion.
Function Versus Expression:

map(): map() requires passing a function as its first argument, which is applied to each element of the iterable. The function can be a built-in function, a lambda function, or a user-defined function.
List comprehensions: List comprehensions use expressions instead of explicit function calls. The expression defines the transformation or evaluation to be applied to each element. It can involve arithmetic operations, function calls, conditional expressions, and more.
