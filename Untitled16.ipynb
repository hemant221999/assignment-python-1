#Q1 To what does a relative path refer?
answer-A relative path is a file path that specifies the location of a file or directory relative to the current working directory of the program or the user's current directory in the file system. It does not include the root directory or the full path to the file or directory.

For example, consider a file example.txt located in the directory my_project. If the current working directory is my_project, the relative path to example.txt would simply be example.txt. However, if the current working directory is my_project/data, the relative path to example.txt would be ../example.txt, where .. refers to the parent directory of the current working directory.

In contrast, an absolute path specifies the complete path to a file or directory from the root directory of the file system. This includes all directories and subdirectories necessary to locate the file or directory, regardless of the current working directory.

Relative paths are commonly used in programming to reference files and directories within a project or program, as they allow for flexibility in moving or copying files to different locations without needing to update the path references. However, absolute paths are necessary when referencing files or directories outside of the project or program's directory structure.


#Q2 What does an absolute path start with your operating system?
answer- The format of an absolute path on an operating system depends on the file system used by the operating system. However, on most modern operating systems, an absolute path typically starts with the root directory symbol.

On Unix-based systems, such as Linux or macOS, the root directory is denoted by a forward slash (/). For example, an absolute path on a Unix-based system might look like this: /home/user/Documents/file.txt

On Windows-based systems, the root directory is typically denoted by a drive letter followed by a colon (:), followed by a backslash (\). For example, an absolute path on a Windows-based system might look like this: C:\Users\user\Documents\file.txt

It's worth noting that file paths may vary depending on the specific file system and operating system in use. Additionally, some programming languages may have different conventions for representing file paths, so it's important to check the documentation of the specific language and operating system being used to ensure that the correct format is being used.

#Q6 What are the three “mode” arguments that can be passed to the open() function?
answer- The open() function in Python is used to open files in different modes. The mode argument determines how the file should be opened and what operations can be performed on it. There are three main mode arguments that can be passed to the open() function:

"r" (read mode): This mode is used when you want to read from a file. It is the default mode, so if you don't specify a mode when opening a file, it will be opened in read mode by default.

"w" (write mode): This mode is used when you want to write to a file. If the file already exists, its contents will be erased and replaced with the new data you write to it. If the file does not exist, a new file will be created.

"a" (append mode): This mode is used when you want to add new data to the end of an existing file. If the file does not exist, a new file will be created. The existing data in the file will not be erased.

#Q7 What happens if an existing file is opened in write mode?
answer- If an existing file is opened in write mode ("w"), its contents will be completely overwritten with the new data that is written to it. If the file doesn't already exist, a new file will be created.
For example, consider the following code snippet:
with open("example.txt", "w") as f:
    f.write("This is some example text.")
if the file example.txt already exists in the current working directory, its contents will be completely overwritten with the string "This is some example text.". If the file doesn't exist, a new file named example.txt will be created in the current working directory with the given text.

#Q8 How do you tell the difference between read() and readlines()?
answer-In Python, read() and readlines() are two methods that can be used to read data from a file object.

read() is used to read the entire contents of a file into a single string. It reads the file from the current position of the file pointer up to the end of the file, or up to a specified number of bytes if a size argument is passed.

readlines(), on the other hand, is used to read the entire contents of a file into a list of strings. Each line of the file becomes a separate string element in the list.

#Q9 What data structure does a shelf value resemble?
answer-In Python, a shelf value is an object that resembles a dictionary data structure. The shelve module provides a simple interface to store and retrieve persistent Python objects. The objects are stored in a key-value pair format, where the keys are strings, and the values can be any picklable Python object.

Here's an example of how to use a shelf:

import shelve

# Open a shelf file for reading and writing
with shelve.open('my_shelf') as shelf:
    # Add some data to the shelf
    shelf['key1'] = {'name': 'John', 'age': 30}
    shelf['key2'] = [1, 2, 3, 4, 5]
    shelf['key3'] = 'hello world'

    # Read data from the shelf
    value1 = shelf['key1']
    value2 = shelf['key2']
    value3 = shelf['key3']

    print(value1)
    print(value2)
    print(value3)
