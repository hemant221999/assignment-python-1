#Q8 In standard expressions, what does the | character mean?
answer- In standard regular expression syntax, the | character (vertical bar or pipe symbol) is used to specify an alternative pattern. It is called the "alternation" or "OR" operator.

The | character separates multiple patterns, any of which can be matched in the input string. For example, the regular expression cat|dog matches either the substring "cat" or the substring "dog" in the input string.

Here's an example code using the | character:
import re

text = "The quick brown fox jumps over the lazy dog."
pattern = r'fox|dog'
matches = re.findall(pattern, text)
print(matches)
The output will be ['fox', 'dog'], because the regular expression pattern fox|dog matches either the substring "fox" or the substring "dog" in the input string.

#Q10 In regular expressions, what is the difference between the + and * characters?
answer- In regular expression syntax, the + and * characters are quantifiers that specify how many times the preceding pattern should match. The main difference between them is:

+ matches one or more occurrences of the preceding pattern.
* matches zero or more occurrences of the preceding pattern.
For example, the regular expression ab+ matches the substring "ab" followed by one or more occurrences of the letter "b". This would match strings like "ab", "abb", "abbb", and so on.

#Q20 answer - To match a number with commas for every three digits, you can use the following regular expression:
^\d{1,3}(,\d{3})*$
Explanation:

^ - matches the start of the string
\d{1,3} - matches 1 to 3 digits
(,\d{3})* - matches zero or more occurrences of a comma followed by exactly 3 digits
$ - matches the end of the string
This regular expression matches any number that starts with 1 to 3 digits, followed by zero or more occurrences of a comma and exactly 3 digits. It does not match numbers with less than 4 digits, or numbers with a comma after only two digits.

Here's an example of using this regular expression in Python:
import re

pattern = re.compile(r'^\d{1,3}(,\d{3})*$')

valid_numbers = ['42', '1,234', '6,368,745']
invalid_numbers = ['12,34,567', '1234']

for num in valid_numbers + invalid_numbers:
    match = pattern.match(num)
    if match:
        print(f'{num} is valid')
    else:
        print(f'{num} is invalid')
