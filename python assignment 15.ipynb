#Q7 Write a generator, genPrimes, that returns the sequence of prime numbers on successive calls to
its next() method: 2, 3, 5, 7, 11, ...
answer - Here's an implementation of the genPrimes generator that returns the sequence of prime numbers on successive calls to its next() method:
def genPrimes():
    """
    Generator that yields the sequence of prime numbers on successive calls to its next() method.
    """
    primes = [2]  # initialize the list of primes with the first prime number
    yield 2  # yield the first prime number
    n = 3  # start testing for primes at 3
    while True:
        is_prime = True  # assume n is prime until proven otherwise
        for p in primes:
            if n % p == 0:  # if n is divisible by a prime number, it's not prime
                is_prime = False
                break
        if is_prime:  # if n is prime, add it to the list of primes and yield it
            primes.append(n)
            yield n
        n += 2  # test only odd numbers for primality (all even numbers except 2 are not prime)
The genPrimes() function initializes a list of primes with the first prime number, 2, and yields it. It then tests successive odd integers starting from 3 for primality. If an integer is prime, it is added to the list of primes and yielded by the generator. If an integer is not prime, the generator moves on to the next integer to test. This process continues indefinitely, generating an infinite sequence of prime numbers on successive calls to next()
