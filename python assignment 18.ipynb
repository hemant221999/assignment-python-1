#Q7 Make a default dictionary called dict_of_lists and pass it the argument list. Make the list
dict_of_lists[&#39;a&#39;] and append the value &#39;something for a&#39; to it in one assignment. Print
dict_of_lists[&#39;a&#39;].

To create a defaultdict called dict_of_lists and initialize it with the list type, then append the value 'something for a' to the list dict_of_lists['a'] in a single assignment, and finally print dict_of_lists['a'], 
from collections import defaultdict

dict_of_lists = defaultdict(list)
dict_of_lists['a'].append('something for a')
print(dict_of_lists['a'])
