#Q1 How do you distinguish between shutil.copy() and shutil.copytree()?
answer-shutil.copy() and shutil.copytree() are functions in the Python Standard Library's shutil module used for copying files and directories, respectively. The key differences between them are:
shutil.copy() is used to copy a single file from one location to another, while shutil.copytree() is used to copy an entire directory and its contents to another location.
shutil.copy() creates a new file in the destination directory with the same name as the original file, while shutil.copytree() creates a new directory in the destination location with the same name as the original directory and copies all files and subdirectories recursively into it.
shutil.copy() preserves the file metadata (such as permissions, timestamps, and ownership), while shutil.copytree() can preserve metadata if the copy_function argument is used.
shutil.copy() can overwrite an existing file in the destination directory, while shutil.copytree() cannot overwrite an existing directory in the destination location.

In summary, shutil.copy() is used to copy a single file, while shutil.copytree() is used to copy an entire directory and its contents.

#Q2 What function is used to rename files??
answer-The os.rename() function is used to rename files in Python.

The os.rename() function takes two arguments: the current name of the file, and the new name for the file. Here is an example:
import os

# current name of the file
old_name = "old_filename.txt"

# new name for the file
new_name = "new_filename.txt"

# rename the file
os.rename(old_name, new_name)

#Q5. Create a programme that searches a folder tree for files with a certain file extension (such as .pdf
or .jpg). Copy these files from whatever location they are in to a new folder.

answer- Here's an example program that searches for files with a given extension (in this case, ".pdf") in a folder and its subfolders, and copies them to a new folder:
import os
import shutil

# set the source folder and destination folder paths
source_folder = '/path/to/source/folder'
destination_folder = '/path/to/destination/folder'

# set the file extension to search for
file_extension = '.pdf'

# create the destination folder if it doesn't already exist
if not os.path.exists(destination_folder):
    os.makedirs(destination_folder)

# recursively search the source folder for files with the given extension
for root, dirs, files in os.walk(source_folder):
    for file in files:
        if file.endswith(file_extension):
            # construct the full path of the source file
            source_file_path = os.path.join(root, file)
            
            # construct the full path of the destination file
            destination_file_path = os.path.join(destination_folder, file)
            
            # copy the file to the destination folder
            shutil.copy(source_file_path, destination_file_path)
            
            print(f"Copied file {source_file_path} to {destination_file_path}")
In this program, we use the os and shutil modules. We first set the paths of the source and destination folders, as well as the file extension we want to search for. Then, we create the destination folder if it doesn't already exist using the os.makedirs() function.

Next, we use the os.walk() function to recursively search the source folder and its subfolders for files with the given extension. For each file found, we construct the full paths of the source and destination files using os.path.join(), and copy the file to the destination folder using shutil.copy(). Finally, we print a message indicating which file was copied to which location.
